```python
import pandas as pd
```

```python
# Reading the csv file
df = pd.read_csv('Downloads/EventHistory.csv')
```

```python
df
```

```python
# Changing the first purchase's value to NO_HISTORY
# Since no purchases were made before the first purchase
df['EVENT_TYPE'].iat[0] = 'NO_HISTORY'
```

```python
# creating a variable for counter
count = 0
# creating a for loop to iterate over all the records
# changes values to UNCONIRMED_HISTORY along with the count of prior purchases
for i in range(1,8):
    
    if df['EVENT_TYPE'].iat[i] == 'PURCHASE' and df['CUSTOMER_ACCOUNT_ID'].iat[i] != 'fraudster@fraud.com':
        count = count + 1
        df['EVENT_TYPE'].iat[i] = 'UNCONFIRMED_HISTORY: ' + str(count)
```

```python
df
```

```python
# creating a variable for counter
count = 0
Date1 = int(df['DATE'][0][5:7])
# For purchases that are older than 90 days
# the event type will be changed to GOOD_HISTORY along with the count of records
# creating a for loop to iterate over all the records
for i in range(2,8):
    Date2 = int(df['DATE'][i][5:7])
    
    if Date2-Date1 > 3:
        count = count + 1
        string = str(count)
        df['EVENT_TYPE'].iat[i] = 'GOOD_HISTORY: ' + string
```

```python
df
```

```python
# creating a variable for counter
count = 0
# creating a for loop to iterate over all the records
# changes the event type to FRAUD_HISTORY: (count of reports) for all the fraud reports
for i in range(1,8):
    
    if df['EVENT_TYPE'][i] == 'FRAUD_REPORT':
        count = count + 1
        string = str(count)
        df['EVENT_TYPE'].iat[i] = 'FRAUD_HISTORY: ' + string
```

```python
df
```

```python
# creating a list to store the index values for records that contain the string
# FRAUD_HISTORY
indexlist = []
for i in range(1,8):
    if search("FRAUD_HISTORY", df['EVENT_TYPE'].iat[i]):
        #df.iat[i].drop(axis=0, inplace=True)
        indexlist.append(i)
```
